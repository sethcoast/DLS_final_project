```python
def generate_dataset(model, noise, label, file_path):
    """
    model: model for which to generate dataset
    noise: input noise vector
    label: input label vector (for class conditional generation)
    filepath: where to save the generated imgs

    This method creates a dataset from the noise vector passed in. This is done
    for the purpose of computing FID and IS scores for our pruned generators
    """
    # iterate over noise, producing and saving 100 samples at a time
    for i in range(int(label.shape[0]/100)):
        start = i*100
        stop = i*100+100
        n = noise[start:stop]
        l = label[start:stop]
        print(start, stop)

        # generate 100 imgs
        with torch.no_grad():
            out = model(n, l).detach().cpu()

        # save the imgs to a file
        for j,img in enumerate(out):
            save_image(denorm(img), '{}/{}.png'.format(file_path,start+j))
```

```python
def load_models_and_generate_datasets(model_path, model_names, data_path, noise, label, device):
    """
    model_path: string. The folder where the pruned models are stored
    model_names: List. A list of strings of the pruned model names
    data_path: where to save the dataset (if does not exist this method will create it)
    noise: input noise vector
    label: input label vector
    device: cpu or gpu (device = torch.device('cuda' if torch.cuda.is_available() else 'cpu'))

    This method will load n models given a list of n model names, and create a 
    10k image dataset for each model
    """
    for model_name in model_names:
        print(model_name)

        # full data path (create it if it doesn't exist)
        full_data_path = '{}{}/'.format(data_path, model_name)
        if not os.path.exists(full_data_path):
            os.makedirs(full_data_path)

        # load model
        model = Generator(image_size=128, conv_dim=32, z_dim=128, c_dim=128, repeat_num=5)
        model.load_state_dict(torch.load('{}{}'.format(model_path, model_name)))
        model.to(device)

        # generate the dataset
        generate_dataset(model, noise, label, full_data_path)
```

```python
# load images
def load_imgs(filepath, max_size=10000):
    img_names = os.listdir(filepath)
    name = '{}/{}'.format(filepath, img_names[0])
    print(name)
    img_shape = np.asarray(Image.open(name)).shape

    imgs = np.zeros(shape=(max_size, img_shape[0], img_shape[1], img_shape[2]))
    
    for i, img_name in enumerate(img_names):
        if i >= max_size:
            break
        imgs[i] = np.asarray(Image.open('{}/{}'.format(filepath, img_name)))
    
    return np.moveaxis(imgs, [1,2,3], [2,3,1])
```

```python
def print_model(model):
    """Print out the network information."""
    num_params = 0
    for p in model.parameters():
        num_params += torch.count_nonzero(p).item()
    #print(model)
    print("The number of parameters: {:.1f} M, ({})".format(num_params/10**6, num_params))
    print(num_params)
    
```

```python
def count_models_params(model_folder):
    models_num_params = {}
    for model_name in os.listdir(model_folder):
        # compose model_path
        model_path = '{}/{}'.format(model_folder, model_name)
        # load model
        model = Generator(image_size=128, conv_dim=32, z_dim=128, c_dim=128, repeat_num=5)
        model.load_state_dict(torch.load(model_path))

        # count the nonzero params
        num_params = 0
        for p in model.parameters():
            num_params += torch.count_nonzero(p).item()
        
        # save
        models_num_params[model_name] = num_params
    
    return models_num_params
```
